```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
df_stun = pd.read_excel(
    '/home/rmfarizky/Project/deteksi_stunting/Prepros1.xlsx',
    )

print('Data downloaded and read into a dataframe!')
```

```python
df_stun.info()
```

```python
df_stun["tgl_lahir"] = pd.to_datetime(df_stun["tgl_lahir"])
```

```python
df_stun['usia_ukur'] = df_stun['tgl_ukur'] - df_stun['tgl_lahir']
```

```python
df_stun['usia_ukur']
```

```python
def days_to_months(td):
    return td.days / 30.4375 

df_stun['usia_ukur(bulan)'] = df_stun['usia_ukur'].apply(days_to_months)
```

```python
df_stun['usia_ukur(bulan)']
```

```python
df_stun['usia_ukur(bulan)'] = df_stun['usia_ukur(bulan)'].round().astype(int)
```

```python
df_stun.drop(columns='lila')
```

```python
df_stun.to_excel('Prepros2.xlsx', index=False)
```

```python
# Hitung jumlah total baris
total_rows = len(df_stun)

# Hitung jumlah 'L' (Laki-laki) dan 'P' (Perempuan) dengan proporsi yang diinginkan
total_laki_laki = int(total_rows * 0.47)  # Misalnya, 40% Laki-laki
total_perempuan = total_rows - total_laki_laki

# Buat array dengan nilai 'L' dan 'P' yang sesuai dengan proporsi
jenis_kelamin = np.array(['L'] * total_laki_laki + ['P'] * total_perempuan)

# Acak urutan elemen dalam array
np.random.shuffle(jenis_kelamin)
```

```python
jenis_kelamin
```

```python
df_stun['j_kel'] = jenis_kelamin
```

```python
df_stun.drop(columns='nama_anak', inplace=True)
```

```python
df_stun.drop(columns='lila', inplace=True)
```

```python
df_stun
```

```python
df_stun.to_excel('Prepros3.xlsx', index=False)
```

```python
df_stun = df_stun[['j_kel', 'tgl_lahir', 'tgl_ukur','usia_ukur(bulan)', 'berat','tinggi']]
```

```python
df_stun['tgl_lahir'] = df_stun['tgl_lahir'].dt.date
df_stun['tgl_ukur'] = df_stun['tgl_ukur'].dt.date
```

```python
df_stun.to_excel('Prepros3.xlsx', index=False)
```
