```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/home/rmfarizky/Project/deteksi_stunting/Dataset/DataLB/STUNTING.xlsx")
```

```python

plt.figure(figsize=(12, 6))
colors = ['gold' if i != 3 else 'orangered' for i in range(len(df))]

# Membuat grafik batang horizontal dengan Seaborn
ax = sns.barplot(x='value', y='nama_data', data=df, palette=colors)

# Menambahkan nilai di dekat grafik
for p in ax.patches:
    ax.annotate(f'{p.get_width():.1f}', (p.get_width() + 1, p.get_y() + p.get_height() / 2), ha='left', va='center')

# Menambahkan label dan judul
plt.xlabel('persen (%)')
plt.xticks(range(0, 45, 5))
plt.ylabel('')
plt.title('Top 8 Prevalensi Balita Stunting Indonesia Berdasarkan Provinsi (2022)',fontsize=17)
plt.grid(axis='x', linestyle='-', alpha=0.1)

# Menampilkan grafik
plt.show()
```
